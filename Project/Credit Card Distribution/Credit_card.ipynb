Calculating and Performing the Statistical Math Summary ::

```python
import numpy as  np
import pandas as pd
df = pd.read_csv("/content/creditcard.csv")
df.head(5)
df['Amount'].describe()
```

```python
### Range
range_value = df['Amount'].max()- df['Amount'].min()
print(range_value)
```

### Calculating the Mean,Quartiles and the Vairance  of the Amount

```python
amount_overall = df['Amount'].tolist()
amount_overall
amount_overall.sort()
print(amount_overall)
```

```python
mean_amount = df["Amount"].sum()/len(amount_overall)
print(mean_amount)
```

```python
### Standard Deviation and the Variance
sum = 0
for i in amount_overall:
  sum+= (i-mean_amount)**2
variance = sum/len(amount_overall)
print(variance)
standard_deviation = variance ** 0.5
print(standard_deviation)
```

```python
#### Calculating the Quartiles :: Q1, Q2, and Q3
#### Position of the Quartiles ::
post1 = (len(amount_overall)+1)/4
post2 = (len(amount_overall)+1)/2
post3 = 3*(len(amount_overall)+1)/4
print(post1,post2,post3)
```

```python
q1 = amount_overall[int(post1)-1]
q1
q2 = amount_overall[int(post2)-1]
q3 = amount_overall[int(post3)-1]
print(q1,q2,q3)
```

Calculating the InterQuartiles::

```python
iqr = q3-q1
print(iqr)
```

Calculating the Lower and Upper Outliers

```python
lower_outliers = q1-1.5*iqr
upper_outliers = q3+1.5*iqr
print(f"The Upper Outliers is {upper_outliers}")
print(f"The Lower Outliers is {lower_outliers}")
```

```python
### Chekcing the OUtliers ::
count = 0
for i in amount_overall:
  if i >= upper_outliers or i <= lower_outliers:
    count += 1
  ### The Total Number of the Outliers
print(f"The total number of the OUtliers is {count}")
```

Skeweness and Kurtotis

```python
### Calculating the Central Moment::
### 1st central Moment
first = 0
for i in amount_overall:
  first += (i-mean_amount)
first_cental = first/len(amount_overall)
print(f"The First Central Moment is {first_cental}") ### Assumed to be Zero :: 0
```

```python
### ### Calculating the Central Moment::
### 2nd central Moment
second = 0
for i in amount_overall:
  second += (i-mean_amount)**2
second_central = second/len(amount_overall)
print(f"The Second Central Moment is {second_central}")
```

```python
### ### Calculating the Central Moment::
### 3rd central Moment
third = 0
for i in amount_overall:
  third += (i-mean_amount)**3
third_central = third/len(amount_overall)
print(f"The Third Central Moment is {third_central}")
```

```python
### ### Calculating the Central Moment::
### 4th central Moment
fourth = 0
for i in amount_overall:
  fourth += (i-mean_amount)**4
fourth_central = fourth/len(amount_overall)
print(f"The First Central Moment is {fourth_central}")
```

```python
if third_central == 0:
  print("Symmetric")
elif third_central > 0:
  print("Right Skewed")
else:
  print("Left Skewed")
```

Conclusion ::
###
The distribution of the data is right skewed (positively skewed). This means that most of the values are concentrated on the left side, with a long tail stretching to the right. In practical terms, there are a few unusually large values pulling the mean to the right of the median.

```python
# Kurtosis
beta2 = fourth_central / (second_central**2)

print("Kurtosis (2):", beta2)
```

```python
# Kurtosis type
if beta2 > 3:
    print("Distribution: Leptokurtic")
elif beta2 < 3:
    print("Distribution: Platykurtic")
else:
    print("Distribution: Mesokurtic")
```

Conclusion::
###
The distribution is leptokurtic, which means it has a sharper peak and heavier tails than a normal distribution. This indicates that the data has a higher chance of extreme values.
